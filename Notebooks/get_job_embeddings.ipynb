```python
import os
import json

def create_entity_ruler_jsonl(df, output_dir="/content/drive/MyDrive/data/", output_file="majors_entity_ruler.jsonl"):
    """
    Creates a JSON Lines (.jsonl) file containing entity ruler patterns from a DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing majors in a column (e.g., 'majors').
        output_dir (str, optional): The directory where the output JSON Lines file will be saved. Defaults to current directory.
        output_file (str, optional): The name of the output JSON Lines file. Defaults to "majors_entity_ruler.jsonl".

    Returns:
        bool: True if the file was created successfully, False otherwise.
    """

    try:
        # Extract unique majors (lowercase for consistency)
        majors = df['Major'].unique().tolist()
        majors = [major.lower() for major in majors]  # Convert to lowercase

        # Create the output directory if it doesn't exist
        os.makedirs(output_dir, exist_ok=True)

        # Construct the full path to the output file
        output_path = os.path.join(output_dir, output_file)

        # Open the file in write mode with 'newline='' (important for jsonl)
        with open(output_path, 'w', newline='') as outfile:
            for major in majors:
                # Write each pattern individually
                pattern = {"label": "MAJOR", "pattern": [{"TEXT": major}]}
                outfile.write(json.dumps(pattern) + '\n')
                pattern = {"label": "MAJOR", "pattern": [{"LOWER": token.lower()} for token in major.split()]}
                outfile.write(json.dumps(pattern) + '\n')

        print(f"Entity ruler patterns saved to {output_path}")
        return True

    except Exception as e:
        print(f"Error creating entity ruler file: {e}")
        return False
```

```python
import re

def extract_education(text):
    education = set()  # Use a set to store unique degrees
    pattern = r"(?i)(?:Bsc|\bB\.\w+|\bM\.\w+|\bPhD\.\w+|\bBachelor(?:'s)?|\bMaster(?:'s)?|\bPhD|\bAssociate(?:'s)?|\bVocational(?:'s)?|\bHigh\sSchool)"
    matches = re.findall(pattern, text)
    for match in matches:
        education.add(match.strip())
    return list(education)  # Convert set back to a list before returning
```

```python
df['degree_required'] = df['text_non_stop'].apply(extract_education)
```
