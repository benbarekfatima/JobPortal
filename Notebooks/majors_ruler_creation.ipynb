#create json entity ruler file

```python
import json

def create_entity_ruler_jsonl(majors_df, output_dir="/content/drive/MyDrive/data/", output_file="majors_entity_ruler_2.jsonl"):
    try:
        # Extract unique majors (lowercase for consistency)
        majors = majors_df.iloc[:, 0].dropna().unique().tolist()  # Extract unique non-null values
        majors = [major.lower() for major in majors]  # Convert to lowercase

        # Open the file in write mode with 'newline='' (important for jsonl)
        with open(output_dir + output_file, 'w', newline='') as outfile:
            for major in majors:
                # Write each pattern individually
                pattern = {"label": "MAJOR", "pattern": [{"TEXT": major}]}
                outfile.write(json.dumps(pattern) + '\n')
                pattern = {"label": "MAJOR", "pattern": [{"LOWER": token.lower()} for token in major.split()]}
                outfile.write(json.dumps(pattern) + '\n')

        print(f"Entity ruler patterns saved to {output_dir + output_file}")
        return True

    except Exception as e:
        print(f"Error creating entity ruler file: {e}")
        return False
```

```python
# Assuming majors_df is your DataFrame containing major data

# Call the function to create the JSON file
success = create_entity_ruler_jsonl(majors_df)

if success:
    print("Entity ruler patterns JSON file created successfully!")
else:
    print("Error creating entity ruler patterns JSON file.")
```
