## create json entity ruler file

```python
import json

def create_entity_ruler_jsonl(df, output_file="skills_entity_ruler.jsonl"):
    """
    Creates a JSON Lines (.jsonl) file containing entity ruler patterns from a DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing skills in a column (e.g., 'skill').
        output_file (str, optional): The name of the output JSON Lines file. Defaults to "skills_entity_ruler.jsonl".

    Returns:
        bool: True if the file was created successfully, False otherwise.
    """

    try:
        # Extract unique skills (lowercase for consistency)
        skills = df.iloc[:, 0].unique().tolist()
        skills = [skill.lower() for skill in skills]  # Convert to lowercase

        # Open the file in write mode with 'newline='' (important for jsonl)
        with open(output_file, 'w', newline='') as outfile:
            for skill in skills:
                # Write each pattern individually
                pattern = {"label": "SKILL", "pattern": [{"TEXT": skill}]}
                outfile.write(json.dumps(pattern) + '\n')
                pattern = {"label": "SKILL", "pattern": [{"LOWER": token.lower()} for token in skill.split()]}
                outfile.write(json.dumps(pattern) + '\n')

        print(f"Entity ruler patterns saved to {output_file}")
        return True

    except Exception as e:
        print(f"Error creating entity ruler file: {e}")
        return False
```

# train spacy

```python
#spacy
import spacy
from spacy.pipeline import EntityRuler
from spacy.lang.en import English
from spacy.tokens import Doc
```

```python
nlp = spacy.load("en_core_web_lg")

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")
# Initialize the entity ruler
ruler = nlp.add_pipe("entity_ruler")
# Read patterns from the JSON file
pattern_file = "/content/skill_ruler.jsonl"
with open(pattern_file, "r") as f:
    patterns = f.readlines()
# Add each pattern to the entity ruler
for pattern_str in patterns:
    pattern_dict = json.loads(pattern_str)
    if pattern_dict.get("pattern", []):  # Check if pattern contains tokens
        ruler.add_patterns([pattern_dict])
```

```python
def get_skills_2vec(text):
    doc = nlp(text)
    myset = []
    subset = []
    for ent in doc.ents:
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    return subset
```

```python
def render_entities(job_ad_text):
    doc = nlp(job_ad_text)
    # Filter entities to only include SKILL entities
    skill_entities = [ent for ent in doc.ents if ent.label_ == "SKILL"]
    if skill_entities:
        # Create spans for the SKILL entities
        spans = [{"start": ent.start_char, "end": ent.end_char, "label": ent.label_} for ent in skill_entities]
        # Render the SKILL entities using displacy
        displacy.render({"text": job_ad_text, "ents": spans, "title": None}, style="ent", manual=True)
        print("\n\n")
    else:
        print("No skill entities found in the text.\n\n")

# Apply the function to each row
jobs['job_ad_text'] = jobs['job_ad_text'].apply(remove_stopwords)
jobs['job_ad_text'] = jobs['job_ad_text'].apply(remove_punctuations)
jobs['job_ad_text'] = jobs['job_ad_text'].str.lower()
jobs['entities'] = jobs['job_ad_text'].apply(render_entities)
```
